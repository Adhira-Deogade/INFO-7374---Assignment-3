```
from bs4 import BeautifulSoup
```

```
from urllib.request import urlopen
from urllib.request import Request
```

```
# specify the url
quote_page = "https://seekingalpha.com/article/4236897-facebook-inc-fb-ceo-mark-zuckerberg-q4-2018-results-earnings-call-transcript?part=single&source=confirmation_funnel_app_promo"
```

```
# query the website and return the html to the variable ‘page’
# page = urlopen(quote_page)
user_agent = 'Mozilla/5.0 (iPhone; CPU iPhone OS 5_0 like Mac OS X) AppleWebKit/534.46'
request = urlopen(Request(quote_page, data=None, headers={'User-Agent': user_agent}))
```

```
# parse the html using beautiful soup and store in variable `soup`
soup = BeautifulSoup(request, "html.parser")
```

```
# Take out the <div> of name and get its value
name_box = soup.find("h1", attrs={"class": "name"})
```

```
# name = name_box.text.strip() # strip() is used to remove starting and trailing
print (name_box)
```
